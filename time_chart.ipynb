```python
#將檔案匯入
import pandas as pd
data = pd.read_excel(r"C:\Users\User\Desktop\workshop\drive-download-20220120T032909Z-001\MilkTea_cleaned.xlsx")
data
```

```python
#將dataframe 的時間資料轉換成時間格式 後續繪圖會用到
from  datetime import datetime, timedelta
list_t = []
def time_dict(df):
    if df["主文/回文"]=="主文":
        t = df["發佈時間"]
        nt = datetime.strptime(t,'%Y/%m/%d %H:%M:%S')
        s = nt.strftime("%Y/%m/%d")
        ns = datetime.strptime(s,'%Y/%m/%d')
        list_t.append(ns)
    
data.apply(time_dict,axis=1)
```

```python
#創建dict 用來儲存各平台
dict_t_fb= {}
dict_t_yt= {}
dict_t_ptt= {}

set_t  = set(list_t)
for i in set_t:
    dict_t_fb[i]=0
    dict_t_yt[i]=0
    dict_t_ptt[i]=0
```

```python
#統計前三大平台個時間的討論熱度
def time_count(df):
    if df["主文/回文"]=="主文":
        t = df["發佈時間"]
        nt = datetime.strptime(t,'%Y/%m/%d %H:%M:%S')
        s = nt.strftime("%Y/%m/%d")
        ns = datetime.strptime(s,'%Y/%m/%d')
        if df["平台"]=="Facebook":
            dict_t_fb[ns] += int(df["討論串總則數"])
        if df["平台"]=="Youtube":
            dict_t_yt[ns] += int(df["討論串總則數"])
        if df["平台"]=="PTT":
            dict_t_ptt[ns] += int(df["討論串總則數"])
data.apply(time_count,axis=1)
```

```python
#將收集到的資料轉換成 dataframe 方便觀看及畫圖
platform_time_fb = pd.DataFrame(dict_t_fb,index=["Popularity"]).T
platform_time_fb = platform_time_fb.sort_index(axis=0)

platform_time_yt = pd.DataFrame(dict_t_yt,index=["Popularity"]).T
platform_time_yt = platform_time_yt.sort_index(axis=0)

platform_time_ptt = pd.DataFrame(dict_t_ptt,index=["Popularity"]).T
platform_time_ptt = platform_time_ptt.sort_index(axis=0)
```

```python
import matplotlib.pyplot as plt
#把前三大的平台的平台熱度顯示出來，因為YT和PTT數據相近因此獨立一張圖給他們兩個作呈現
plt.figure(figsize=(16,5))
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.plot(platform_time_fb.index, platform_time_fb["Popularity"])
plt.plot(platform_time_yt.index, platform_time_yt["Popularity"])
plt.plot(platform_time_ptt.index, platform_time_ptt["Popularity"])
plt.title('平台熱度')
plt.legend(["FB","YT","PTT"])
plt.show()

#呈現PTT和YT的時間熱度圖
plt.figure(figsize=(16,5))
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.plot(platform_time_yt.index, platform_time_yt["Popularity"])
plt.plot(platform_time_ptt.index, platform_time_ptt["Popularity"])
plt.title('平台熱度')
plt.legend(["YT","PTT"])
plt.show()
```
