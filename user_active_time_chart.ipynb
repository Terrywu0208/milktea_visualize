```python
import pandas as pd
data = pd.read_excel(r"C:\Users\User\Desktop\workshop\drive-download-20220120T032909Z-001\MilkTea_cleaned.xlsx")
data
```

```python
#抓取時間
data["datetime"]=pd.to_datetime(data["發佈時間"])
data["time"]=data["datetime"].apply(lambda x :x.strftime("%H"))
timecat=data["time"].sort_values()
```

```python
timecat.value_counts().sort_index()
```

```python
#繪圖
import matplotlib.pyplot as plt
import numpy as np
plt.bar(timecat.unique(),#把一樣時間的資料歸類(np.array型態)
        timecat.value_counts().sort_index(),#把各種時間類別做計算(np.array型態)
        width=0.5, 
        bottom=None, 
        align='center', 
        color=['royalblue'],)
plt.xticks(rotation='horizontal')
plt.xlabel("幾點(24小時制)")
plt.ylabel("討論總熱度")
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']#畫中文字型一定要這一行!
plt.show()
```
