```python
#將檔案匯入
import pandas as pd 
data = pd.read_excel(r"C:\Users\User\Desktop\workshop\drive-download-20220120T032909Z-001\MilkTea_cleaned.xlsx")
data
```

```python
data.info()
```

```python
#收集有哪些平台並創建dict
platform_kind =  set(data["平台"])
platform_dict = {}
for i in platform_kind:
    platform_dict[i]=0
platform_dict["Other(except:FB)"]=0
platform_dict["Other(except:FB & Instagram & Youtube & PTT)"]=0
platform_dict 
```

```python
#計算各平台的討論串總則數以當作熱度依據
def platform_count(df):
    if df["主文/回文"]=="主文":
        platform_dict[df["平台"]] += int(df["討論串總則數"])
        if df["平台"]!="Facebook":
            platform_dict["Other(except:FB)"] += int(df["討論串總則數"])
        if df["平台"]!="Facebook" and df["平台"]!="Youtube" and df["平台"]!="PTT" and df["平台"]!="Instagram":
            platform_dict["Other(except:FB & Instagram & Youtube & PTT)"] += int(df["討論串總則數"])

data.apply(platform_count,axis=1)
```

```python
#將dict轉變成dataframe 方便觀看和畫圖
platform_df = pd.DataFrame(platform_dict,index=["Popularity"]).T
platform_df = platform_df.sort_values(by=["Popularity"],ascending = False)
platform_df_display = platform_df.drop(["Other(except:FB)","Other(except:FB & Instagram & Youtube & PTT)"],axis=0)
platform_df_display
```

```python
#分析個平台熱門程度(以圓餅圖呈現)
import matplotlib.pyplot as plt
label_list = platform_df.index[:2]
color_list = []
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.pie(platform_df["Popularity"][:2],labels = label_list,autopct='%1.2f%%')
plt.title('平台熱度')
plt.show()
```

```python
#分析個平台熱門程度(以圓餅圖呈現) 剔除FB觀察其他平台的熱門程度
import matplotlib.pyplot as plt
label_list = platform_df.index[2:6]
color_list = []
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.pie(platform_df["Popularity"][2:6],labels = label_list,autopct='%1.2f%%')
plt.title('平台熱度(除了主要的FB以外)')
plt.show()
```
