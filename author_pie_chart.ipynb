```python
#將檔案匯入
import pandas as pd 
data = pd.read_excel(r"C:\Users\User\Desktop\workshop\drive-download-20220120T032909Z-001\MilkTea_cleaned.xlsx")
data
```

```python
data.info()
```

```python
#收集有哪些平台並創建dict
author_kind =  set(data["作者"])
author_dict = {}
for i in author_kind:
    author_dict[i]=0
author_dict 
```

```python
#計算各平台的討論串總則數以當作熱度依據
def author_count(df):
    if df["主文/回文"]=="主文":
        author_dict[df["作者"]] += int(df["討論串總則數"])
data.apply(author_count,axis=1)
```

```python
#將dict轉變成dataframe 方便觀看和畫圖
author_df = pd.DataFrame(author_dict,index=["Popularity"]).T
author_df = author_df.sort_values(by=["Popularity"],ascending = False)
author_df
```

```python
#分析個平台熱門程度(以圓餅圖呈現)
import matplotlib.pyplot as plt
label_list = author_df.index[:6]
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.pie(author_df["Popularity"][:6],labels = label_list,autopct='%1.2f%%')
plt.title('作者熱度')
plt.show()
```
