```python
#將檔案匯入
import pandas as pd 
data = pd.read_excel(r"C:\Users\User\Desktop\workshop\drive-download-20220120T032909Z-001\MilkTea_cleaned.xlsx")
data
```

```python
data.info()
```

```python
#收集有哪些平台並創建dict
channel_kind =  set(data["頻道"])
channel_dict = {}
for i in channel_kind:
    channel_dict[i]=0
channel_dict 
```

```python
#計算各平台的討論串總則數以當作熱度依據
def channel_count(df):
    if df["主文/回文"]=="主文":
        channel_dict[df["頻道"]] += int(df["討論串總則數"])
data.apply(channel_count,axis=1)
```

```python
#將dict轉變成dataframe 方便觀看和畫圖
channel_df = pd.DataFrame(channel_dict,index=["Popularity"]).T
channel_df = channel_df.sort_values(by=["Popularity"],ascending = False)
channel_df.head(10)
```

```python
#分析個平台熱門程度(以圓餅圖呈現)
import matplotlib.pyplot as plt
label_list = channel_df.index[:6]
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.pie(channel_df["Popularity"][:6],labels = label_list,autopct='%1.2f%%')
plt.title('頻道熱度')
plt.show()
```
